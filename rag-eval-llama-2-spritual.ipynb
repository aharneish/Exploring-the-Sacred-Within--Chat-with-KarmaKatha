```python
from  llama_index.core.evaluation import FaithfulnessEvaluator,RelevancyEvaluator,CorrectnessEvaluator,AnswerRelevancyEvaluator
from  llama_index.core.evaluation import BatchEvalRunner

ans_rel=AnswerRelevancyEvaluator(service_context=service_context)
faith_rel=FaithfulnessEvaluator(service_context=service_context)
rel=RelevancyEvaluator(service_context=service_context)
runner = BatchEvalRunner({"faithfulness": faith_rel, "relevancy": rel,"answer":ans_rel},workers=8,)
```

```python
def get_eval_results(key, eval_results):
    results = eval_results[key]
    correct = 0
    for result in results:
        if result.passing:
            correct += 1
    score = correct / len(results)
    print(f"{key} Score: {score}")
    return score
```

def evaluate_query_engine(query_engine, questions):
    c = [query_engine.aquery(q) for q in questions]

    print("finished query")

    for r in results:
        # evaluate with gpt 4
        eval_result = (
            1 if faith_rel.evaluate_response(response=r).passing else 0
        )
        total_correct += eval_result
