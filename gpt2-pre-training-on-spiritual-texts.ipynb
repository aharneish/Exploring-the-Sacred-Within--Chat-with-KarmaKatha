```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer, GPT2Config
from transformers import TextDataset, DataCollatorForLanguageModeling
from transformers import Trainer, TrainingArguments
#from transformers import AutoTokenizer, AutoModelForCausalLM
import os
import torch
from transformers import AutoTokenizer,AutoModelForCausalLM
from datasets import load_dataset
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
dataset="Aharneish/spirit"
```

```python
dataset = load_dataset(dataset,split="validation")
```

```python
dataset
```

```python
model=AutoModelForCausalLM.from_pretrained("gpt2-medium")
tokenizer=AutoTokenizer.from_pretrained("gpt2-medium")
```

```python
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

```python
training_args = TrainingArguments(
    output_dir="gpt-2-medium-train-1",
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=4,
    learning_rate=1e-5,
    save_steps=500,
    save_total_limit=2,
    push_to_hub=True,
    evaluation_strategy="steps",  # Evaluate every 'eval_steps' steps
    #eval_steps=500,
    save_strategy="steps",
    #load_model_at_startup=True,
    run_name="gpt-2-medium-train-1",
    fp16=True
)
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=dataset,
    #eval_dataset=eval_data,
    tokenizer=tokenizer
)
```

```python
trainer.train()
trainer.push_to_hub()
```
